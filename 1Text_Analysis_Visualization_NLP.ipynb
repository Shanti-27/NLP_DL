# Vizualize the text

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_stop = set(stopwords.words('english'))
word_stop
```

```python
wordcloud = WordCloud(max_words=100, stopwords=word_stop)
wordcloud.generate(complete_paragraph)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,14))
plt.imshow(wordcloud)
plt.show()
```

```python
word_stop = list(word_stop)  # Convert set to list
wordcloud = WordCloud(max_words=100, stopwords=word_stop + ['phone', 'phones', 'nokia', 'mobile', 'get', 'key'])
wordcloud.generate(complete_paragraph)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,14))
plt.imshow(wordcloud)
plt.show()
```

```python
# changing background color

wordcloud = WordCloud(max_words=100, stopwords=word_stop, background_color = 'yellow')
wordcloud.generate(complete_paragraph)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,14))
plt.imshow(wordcloud)
plt.show()
```

# Regular Expression

```python
import re
```

```python
words = re.split(r'\W+', text)
print(words)
```

```python
word = text.split()
word
```

```python
import re
import string

# compile all together
words = text.split()
re_puct = re.compile('[%s]' %re.escape(string.punctuation))
original = [re_puct.sub('',w) for w in words]
print(original)
```

```python
re_puct = re.compile('[^%s]' %re.escape(string.printable))
# ^ - hat/cap : Consider
# % : calling
# s : symbol
original = [re_puct.sub('',w) for w in words]
print(original)
```
